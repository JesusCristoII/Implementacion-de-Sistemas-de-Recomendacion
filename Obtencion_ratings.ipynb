```python
from rs_datasets import Amazon
import ssl
import urllib.request

# Crea un contexto SSL que no verifica certificados
context = ssl._create_unverified_context()

# Reemplazar el opener global de urllib para que use el contexto SSL inseguro
opener = urllib.request.build_opener(urllib.request.HTTPSHandler(context=context))
urllib.request.install_opener(opener)

amazon = Amazon('cards')
amazon.info()
```

```python
amazon.ratings.to_csv('ratings_practica2.csv')
```

```python
import pandas as pd

df = pd.read_csv('ratings_practica2.csv', index_col= 0)
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['item_id'] = le.fit_transform(df['item_id'])
df['user_id'] = le.fit_transform(df['user_id'])
```

```python
df
```

```python
df.drop_duplicates(subset= ['user_id', 'item_id'], inplace= True)
```

```python
df.to_csv('ratings_bien.csv', index= False)
```
